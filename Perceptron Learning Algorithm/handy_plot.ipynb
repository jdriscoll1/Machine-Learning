```python
# plot an implicite function. 

# Sometimes it's hard or tedious to get dots of an equation, for plotting.
# 
# If a function is given in the form of f(x1,x2)=0, which is called an implicite function, 
# calculating/get dots of this function for plotting is tedious or sometime impossible.

# For ex.: once you find the w = [w0, w1, w2]. The cutting line you want to 
# plot is w0 + w1 * x1 + w2 * x2 = 0. It is possible to get points 
# from this equation, because it's linear and easy to solve. But if f(x1,x2) 
# is not linear, sovling the equation is not even possible, sometimes.

# That's when tools for plotting implicite function kick in to help. 
# These tools use numerical analysis algorithms to find points for the 
# given implicite equation (approximately, if they have to)


delta = 0.01
xrange = np.arange(-20.0, 20.0, delta)
yrange = np.arange(-20.0, 20.0, delta)
X, Y = np.meshgrid(xrange,yrange)
plt.xlabel('X')
plt.ylabel('Y')

# this is to plot the function F(X,Y) = 0.3 + 0.4 X^3 + 10 XY + 4 X^2Y + 2 XY^2 = 0
F = 0.3 + 0.4*X*X*X + 10*X*Y + 4* X* X * Y + 2 * X*Y*Y 


plt.contour(X, Y, (F), [0])  #this is to plot function F = 0
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng
import time


def f(x):
    return 2 * x 
# This creates the graphs line 
x = np.linspace(0,1, 100)
y = f(x)

# alright, next let's create w - it's going to be a 3x1 array w/ 3 random values
# it'll determine the equation w[0] + w[1]*x[1] + w[2]x[2]
x_range = np.arange(0, 1, .01)
y_range = np.arange(0, 1, .01)
x1, x2 = np.meshgrid(x_range,y_range)
w = default_rng(int(time.time())).random((3,1))
F = w[0] + x1*w[1] + x2*w[2]



# Let's slap down a chunk of x's 
points = default_rng(int(time.time())).random((100,2))

pos = np.empty([1,2])
neg = np.empty([1,2])

for i in range(100):
    if f(points[i][0]) < points[i][1]:

        pos = np.insert(pos, 0, points[i], 0)
    
    else: 

        neg = np.insert(neg, 0, points[i], 0)
        
# delete the init/useless row which is on the bottom        
pos = np.delete(pos, -1, 0)
neg = np.delete(neg, -1, 0)


# display the data samples and the target classifier that PLA is tring to learn
plt.scatter(pos[:,0], pos[:,1], marker='+')
plt.scatter(neg[:,0], neg[:,1], marker='_')


# This puts the line on the graph 
plt.plot(x, F[0], '-r', label='y=2x+1')


# This makes the graph's properties 
plt.title('Graph of y=2x+1')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')

# This sets the limits of the graph 
plt.xlim([0, 1])
plt.ylim([0, 1])

plt.contour(x1, x2, F, [0], color='red')

# This outputs the graph 
plt.show()

```
